```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset with error handling
file_path = 'USvideos.csv'
try:
    df = pd.read_csv(file_path)
    print("File loaded successfully.")
except FileNotFoundError:
    print(f"File not found. Please check the file path: {file_path}")
    # If running in an environment where you can upload files, use that option here
    # or prompt the user to provide the correct path.
    raise

# Display DataFrame info and statistics
print(df.info())
print(df.describe())

# Identify numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Plot histograms and KDEs for each numerical column
plt.figure(figsize=(16, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# Plot boxplots for each numerical column
plt.figure(figsize=(16, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
# Calculate and plot the correlation matrix
numerical_df = df[numerical_cols]
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plot pairplots for the numerical columns
sns.pairplot(numerical_df)
plt.show()
```
