```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("householdtask3.csv")
data.head(10)
```

```python
df=pd.DataFrame(data)
df
```

```python
df.set_index('year', inplace=True)
```

```python
# Create a bar chart for 'tot_hhs'
plt.figure(figsize=(8,4))
plt.bar(df.index, df['tot_hhs'], color='blue', label='Total Households')
plt.xlabel('Year')
plt.ylabel('Total Households')
plt.title('Total Households Over the Years')
plt.legend()
plt.show()
```

```python
# Create a line chart for 'income' and 'expenditure'
plt.figure(figsize=(8,4))
plt.plot(df.index, df['income'], marker='o', linestyle='-', color='green', label='Income')
plt.plot(df.index, df['expenditure'], marker='x', linestyle='--', color='red', label='Expenditure')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Income vs Expenditure Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```
